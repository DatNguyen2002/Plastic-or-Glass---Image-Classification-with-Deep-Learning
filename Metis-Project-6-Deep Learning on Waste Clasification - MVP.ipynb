# Garebage Inteligence
## Metis Project 6 - Deep Learning for Plastic/Glass Waste Classification - MVP
Krystian Krystkowiak 2022

The goal of this project is building of **deep learning model** capable of **classifying glass and plastic  waste.** 

It could be part of larger waste classifying model. Focus is put on glass and plastic as it may be interesting challenge.

**Sollution Path:**

I'm using the dataset that originally comes from: https://www.kaggle.com/datasets/mostafaabla/garbage-classification

**2876 images** belonging to **4 classes**: plastic (865 images), green-glass (629 images), brown-glass (607 images) and white-glass (775 images).

**Work Completed:**

After inicial EDA and cleaning, I pre-processed the data and loaded them to **data frame** (in unified resolution **128x128, rgb**).

Next, I created PCA visualisation to visualise **classes separability**.

Next,I created **Logistic Regression** model to have **baseline** for Deep Learning model's performance:

Training Accuracy: 	0.494 | Training F1 score: 	0.466 

Validation Accuracy: 	0.467 | Validation F1 score: 	0.446

**Moving Forward:**

I will proceed to building a Neural Network Model, Convolutional Neural Networks and Transfer Learning to improve F1 performance of the model.
